# Pickled weights, losses, accuracies

```python
with open("projection_acas.pickle", "rb") as f:
    data = pickle.load(f)
```

```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
```

```python
all_accuracies = np.array([[a.numpy() for a in d] for d in data["accuracies"]]).flatten()
```

```python
plt.figure()
plt.title("Loss vs Batch")
plt.xlabel("Batch")
plt.ylabel("Loss")
plt.plot(np.array(data["losses"]).flatten())
plt.ylim([0, 100000])
plt.show()
```

```python
plt.figure()
plt.title("Loss vs Batch")
plt.xlabel("Batch")
plt.ylabel("Loss")
plt.plot(moving_average(np.array(data["losses"]).flatten(), 2**8))
plt.ylim([0, 1000])
plt.show()
```

```python
plt.figure()
plt.title("Accuracy vs Batch")
plt.plot(all_accuracies)
plt.xlabel("Batch")
plt.ylabel("Accuracy")
plt.ylim([0, 1.0])
plt.show()
```
